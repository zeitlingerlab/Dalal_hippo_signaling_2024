## Helper functions

````python
def onehotencode(s):
    """
    One hot encode the sequence to a 4xL numpy array for ACGT 
    """
    l_ = len(s)
    encode = np.zeros((4, l_), dtype = np.int8)
    s = s.upper()
    if len(s.lstrip('ACGT')) == 0:
        encode[np.array(list(map('ACGT'.index, s))), np.arange(l_)] = 1
        return encode
    raise ValueError('Invalid sequence')


def populate_MS_sequences(motifs,scores):
    """
    Populate the sequences based on the motifs and scores by repeating the motifs based on the scores
    """
    sequences = []
    for motif, score in zip(motifs, scores):
        score = int(score)
        for i in range(score):
            sequences.append(motif)
    return sequences



def pwm_score(sequences, pseudocount=1e-6):
    """
    Calculate the PWM score for the sequences
    """
    pwm_matrix = np.zeros((4, len(sequences[0])))
    for sequence in sequences:
        for pos, nucleotide in enumerate(sequence):
            if nucleotide == 'A':
                pwm_matrix[0, pos] += 1
            elif nucleotide == 'C':
                pwm_matrix[1, pos] += 1
            elif nucleotide == 'G':
                pwm_matrix[2, pos] += 1
            elif nucleotide == 'T':
                pwm_matrix[3, pos] += 1
    pwm_matrix += pseudocount
    
    pwm_matrix /= (len(sequences) + 4 * pseudocount) 
    # background_prob = 0.25
    # pwm_matrix = np.log2(pwm_matrix / background_prob)
    pwm_matrix = pwm_matrix - np.mean(pwm_matrix, axis=0)
    pwm_matrix[pwm_matrix < 0] = 0 
    return pwm_matrix
````

## Data preparation

````python
tead4_bp17_sequences,tead4_bp17_scores = tead4_bp17['sequence'], tead4_bp17['tead4']
tead4_bp9_sequences,tead4_bp9_scores = tead4_bp9['sequence'], tead4_bp9['tead4']
````

````python
populated_tead4_bp17_sequences = populate_MS_sequences(tead4_bp17_sequences, tead4_bp17_scores)
populated_tead4_bp9_sequences = populate_MS_sequences(tead4_bp9_sequences, tead4_bp9_scores)

tead4_bp17_pwm = pwm_score(populated_tead4_bp17_sequences)
tead4_bp9_pwm = pwm_score(populated_tead4_bp9_sequences)
````
